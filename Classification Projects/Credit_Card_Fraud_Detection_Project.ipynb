```python
# Removing the last column which is containing the NaN value.
df=credit_card_data.drop(11958)
df.tail()
```

```python
# Distribution of legit transaction and fraudulent transactions.
credit_card_data['Class'].value_counts()
```

```python
df.isnull().sum()
```

This Dataset is Highly unbalanced.

0 --> Represent Normal Transaction.

1 --> Represent Fraudulent Transaction.

```python
# Separating the data for analysis.
legit=df[df.Class==0]
fraud=df[df.Class==1]
```

```python
legit.shape
```

```python
fraud.shape
```

```python
# Stastistical Measures of the data.
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# Compare the values for both Transactions.
df.groupby('Class').mean()
```

Under-Sampling :

Build a sample dataset containing similar distribution of normal transactions and fraudulent transactions.

Number of fraudulent transactions = 52

```python
legit_sample=legit.sample(n=52)
```

Concatenating two DataFrames.

```python
new_df=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_df.head()
```

```python
new_df.tail()
```

```python
new_df['Class'].value_counts()
```

```python
new_df.groupby('Class').mean()
```

Splitting the data into Features and **Target**

```python
features=new_df.drop(columns='Class',axis=1)
target=new_df['Class']
```

```python
features.head()
```

```python
target.head()
```

***Split the data into training and testing data***

```python
features_train,features_test,target_train,target_test=train_test_split(features,target,test_size=0.2,stratify=target,random_state=2)
```

```python
features.shape,features_train.shape,features_test.shape
```

# ***Model Training***

```python
model=LogisticRegression()
```

Training the dataset

```python
model.fit(features_train,target_train)
```

Model Evaluation

```python
# Accuracy Score on Training Data.
training_data_prediction=model.predict(features_train)
training_data_accuracy=accuracy_score(training_data_prediction,target_train)
print('Accuracy on Training Data : ',training_data_accuracy)
```

```python
# Accuracy Score on Test Data.
test_data_prediction=model.predict(features_test)
test_data_accuracy=accuracy_score(test_data_prediction,target_test)
print('Accuracy on Test Data : ',test_data_accuracy)
```
